```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/raw/demand_data.csv")

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])

# Aggregate demand across all stores & departments
daily_demand = (
    df.groupby('Date')['Weekly_Sales']
    .sum()
    .reset_index()
)

# Plot historical demand
plt.figure(figsize=(12,5))
plt.plot(daily_demand['Date'], daily_demand['Weekly_Sales'])
plt.title("Historical Retail Demand (Walmart)")
plt.xlabel("Date")
plt.ylabel("Weekly Sales (Demand)")
plt.grid(True)
plt.show()
```
